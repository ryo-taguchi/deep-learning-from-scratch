```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 6, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle = "--", label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def linear_function(x):
    return np.array(-x + 0.5)

x = np.arange(-0.5, 1.1, 0.1)
y = linear_function(x)
plt.plot(x, y)
plt.ylim(-0.5, 1.1)
plt.show()
```
